```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def magnetization(t):
    return (1-(np.sinh(2/t)**(-4)))**(0.125)
```

```python
import pandas as pd
data = pd.read_csv('ising_magnetization_data_wm100_img100_2')
data_labels=np.array(data['0'])
temp = data_labels[0::100]
data_wo_labels= data.drop(['Unnamed: 0','0'],axis=1)
data_in_array = np.array(data_wo_labels)

mcmc_mag_data=[]
mag=0
for i in range(len(data_in_array)+1):
  

  if i % 100 !=0 or i==0 :
    x_decoded = np.array(data_in_array[i])
    mag=np.abs(np.sum((2*x_decoded - 1)/(28*28))) + mag
  else:
    mcmc_mag_data.append(mag/100)
    mag=0
```

```python
temp = np.linspace(0.01,5,50)
labels=["Metropolis-Hasting Algorithm","Onsenger's curve","z = 2", "z = 3", "z = 4", "z = 5"]
mag_data = [mag_data_conv2,mag_data_conv3,mag_data_conv4,mag_data_conv5]
markers = ["-s","-p","-*","-h","-H"]
tcritical = 2.269
x_data = list(np.arange(.05,5,.01))


y_data = [magnetization(t) if t<tcritical else 0 for t in x_data]


plt.plot(x_data,y_data,"-")
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.grid(True)
ax.plot(temp,mcmc_mag_data,'-x')
ax.plot(x_data,y_data,linewidth=2,alpha=0.8)

ax.legend(labels)
ax.set_xlim(0.01,5.1)
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("Average Magnetization per lattice point")
ax.set_ylim(0,1.10)
ax.set_title("Benchmark for Magnetization per spin from Ising Models \n by Metropolis-Hasting Algorithm and Onsenger's formula")
fig.savefig("benchmark")
```
